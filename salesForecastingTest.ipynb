```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```python
store_sales = pd.read_csv("train.csv")
store_sales.head(10)
```

Check for null values in the dataset

```python
store_sales.info()
```

Dropping Order ID, Product, Quantity Ordered, Price Each, Purchase Address

```python
#store_sales = store_sales.drop(['Order ID', 'Product', 'Price Each', 'Purchase Address'], axis=1)
store_sales = store_sales.drop(['store', 'item'], axis=1)
```

```python
store_sales.info()
```

```python
store_sales['Quantity Ordered'] = pd.to_numeric(store_sales['Quantity Ordered'], errors='coerce')
# Drop rows with NaN
store_sales = store_sales.dropna(subset=['Quantity Ordered'])
```

```python
store_sales['Quantity Ordered'] = pd.to_numeric(store_sales['Quantity Ordered'], errors='coerce').astype('Int64')
```

coverting Order Date from object datatype to date time datatype and Quantity Ordered from object datatype to int64

```python
store_sales['Order Date'] = pd.to_datetime(store_sales['Order Date'], errors='coerce')
store_sales.info()
```

converting date to a Month period, and then sum the nmber of items in each month

```python
store_sales['Order Date'] = store_sales['Order Date'].dt.to_period("M")
monthly_sales = store_sales.groupby('Order Date').sum().reset_index()
```

convert the resulting date column to timestamp datatype

```python
monthly_sales['Order Date'] = monthly_sales['Order Date'].dt.to_timestamp()
```

```python
monthly_sales.head(12)
```

Visualization

```python
plt.figure(figsize=(20,5))
plt.plot(monthly_sales['Order Date'], monthly_sales['Quantity Ordered'])
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Customer Sales")
plt.show()
```

call the difference on the sales columns to make the sales data stationery // march's quantity ordered - feb's quantity ordered, 17005 - 13449 = 2587

```python
monthly_sales['sales_diff'] = monthly_sales['Quantity Ordered'].diff()
monthly_sales = monthly_sales.dropna()
monthly_sales.head(10)
```

```python
plt.figure(figsize=(15,5))
plt.plot(monthly_sales['Order Date'], monthly_sales['Quantity Ordered'])
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Customer Sales Difference")
plt. show()
```

Dropping off sales and date

```python
supervised_data = monthly_sales.drop(['Order Date', 'Quantity Ordered'], axis=1)
```

preparing the supervised data

```python
for i in range(1,13):
    col_name = 'month_' + str(i)
    supervised_data[col_name] = supervised_data['sales_diff'].shift(i)
supervised_data = supervised_data.dropna().reset_index(drop=True)
supervised_data.head(5)
```

Split the data into Train and Test

```python
train_data = supervised_data[:-12]
test_data = supervised_data[-12:]
print("Train Data Shape: ", train_data.shape)
print("Test Data Shape: ", test_data.shape)
```

```python
scaler = MinMaxScaler(feature_range=(-1,1))
scaler.fit(train_data)
train_data = scaler.transform(train_data)
test_data = scaler.transform(test_data)
```

```python
X_train, y_train = train_data[ : ,1:], train_data[: , 0:1]
X_test, y_test = test_data[:,1:], test_data[:,0:1]
y_train = y_train.ravel()
y_test = y_test.ravel()
print("X_train Shape: ", X_train.shape)
print("y_train Shape: ", y_train.shape)
print("X_test Shape: ", X_test.shape)
print("y_test Shape: ", y_test.shape)
```

Make prediction data frame to merge the predicted sales prices of all trained algs

```python
sales_dates = monthly_sales['Order Date'][-12:].reset_index(drop=True)
predict_df = pd.DataFrame(sales_dates)
```

```python
act_sales = monthly_sales['Quantity Ordered'][-13:].to_list()
print(act_sales)
```

To create the linear regression model, and predicted output

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
lr_pre = lr_model.predict(X_test)
```

```python
lr_pre = lr_pre.reshape(-1,1)
# This is a set matrix - contains the input features of the test data, and also the predicted output
lr_pre_test_set = np.concatenate([lr_pre, X_test], axis=1)
lr_pre_test_set = scaler.inverse_transform(lr_pre_test_set)
```

```python
result_list =[]
for index in range(0, len(lr_pre_test_set)):
  result_list.append(lr_pre_test_set[index][0] + act_sales[index])
lr_pre_series = pd.Series(result_list, name="Linear Prediction")
predict_df = predict_df.merge(lr_pre_series, left_index = True, right_index= True)
```

```python
# print(predict_df)
lr_mse = np.sqrt(mean_squared_error(predict_df['Linear Prediction' ], monthly_sales['Quantity Ordered' ] [-12: ]) )
lr_mae = mean_absolute_error(predict_df['Linear Prediction'], monthly_sales['Quantity Ordered' ] [-12: ])
lr_r2 = r2_score(predict_df['Linear Prediction' ], monthly_sales['Quantity Ordered' ] [-12:])
print("Linear Regression MSE: ", lr_mse )
print("Linear Regression MAE: ", lr_mae )
print("Linear Regression R2: ", lr_r2 )
```

```python
plt.figure(figsize=(15,5))
# Actual sales
plt.plot(monthly_sales['Order Date'], monthly_sales['Quantity Ordered'])
# Predicted sales
plt.plot(predict_df['Order Date'], predict_df['Linear Prediction' ])
plt.title("Customer sales Forecast using LR Model")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend(['Actual Sales', 'Predicted sales'])
plt.show()
```
